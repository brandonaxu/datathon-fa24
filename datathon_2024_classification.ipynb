```python
from keybert import KeyBERT
from sentence_transformers import SentenceTransformer
```

```python
model = SentenceTransformer("all-MiniLM-L6-v2")
kw_model = KeyBERT(model=model)
```

```python
def embedding_generator(text):
  doc_embeddings = kw_model.extract_embeddings(text)[0]
  return doc_embeddings
```

```python
news_titles = pd.read_csv('Conflict_Likelihood_Dataset.csv')
```

```python
news_titles.head()
```

```python
corpus = []
conflict_vals = news_titles['Likelihood of Conflict'].tolist()

for idx,desc in news_titles.iterrows():
    text = desc['Text']
    val = desc['Likelihood of Conflict']
    desc_embedding = embedding_generator(text)
    corpus.append(desc_embedding)
```

```python
import heapq

def conflict_value(text, k=5):
    k_closest = []
    k_conflict_vals_indices = []
    heapq.heapify(k_closest)
    embedding = embedding_generator(text)
    for idx,vec in enumerate(corpus):
        dist = np.linalg.norm(embedding-vec)
        heapq.heappush(k_closest, (-dist,idx))
        if len(k_closest)>k:
            heapq.heappop(k_closest)
    vals = [conflict_vals[idx] for val, idx in k_closest]
    return np.mean(vals)
```

```python
def country_live_news(country, weeks=1):
    current_date = datetime.now().date()
    two_weeks_ago = current_date - timedelta(weeks=weeks)
    # current_date_str = current_date.strftime("%Y-%m-%d")
    # two_weeks_ago_str = two_weeks_ago.strftime("%Y-%m-%d")
    nyt_articles = nyt.article_search(
    query=country,
    dates={"begin": two_weeks_ago, "end": current_date},
    results=20
    )
    nyt_titles = []
    nyt_snippets = []
    nyt_dates = []
    nyt_url = []
    
    for article in nyt_articles:
        nyt_titles.append(article['headline']['main'])
        nyt_snippets.append(article.get('snippet', 'No snippet available'))
        nyt_dates.append(article['pub_date'])
        nyt_url.append(article['web_url'])
        
    conflict_vals = []
    
    for idx,title in enumerate(nyt_titles):
        given_date_str = str(nyt_dates[idx])
        given_date = datetime.fromisoformat(given_date_str)
        difference = datetime.now(timezone.utc) - given_date
        days_difference = difference.days
        weight = 7*weeks+1-days_difference
        conflict_val = conflict_value(title)
        conflict_vals += [conflict_val]*weight

    current_date = datetime.now() 
    week_ago_date = current_date - timedelta(days=weeks*7)
    current_date_formatted = current_date.strftime("%Y-%m-%d")
    week_ago_date_formatted = week_ago_date.strftime("%Y-%m-%d")

    f = Filters(
    keyword = country,
    start_date = week_ago_date_formatted,
    end_date = current_date_formatted,
    num_records = 10,
    country = fips_codes['Ukraine']
    )
    articles = gd.article_search(f)
    articles['date'] = articles['seendate'].apply(date_converter)
    articles['translated'] = articles['title'].apply(lambda x: translator.translate(x, src='auto', dest='en').text)

    for idx,row in articles.iterrows():
        date = row['date']
        given_date = datetime.strptime(date, "%m-%d-%Y")
        days_difference = (current_date - given_date).days
        weight = 7*weeks+1-days_difference
        conflict_val = conflict_value(row['translated'])
        conflict_vals += [conflict_val]*weight
    
    return np.mean(conflict_vals)
```
